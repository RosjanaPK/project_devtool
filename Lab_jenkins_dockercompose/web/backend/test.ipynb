```python
import os
from flask import Flask, render_template
import cv2
import requests

import base64
from fastapi import FastAPI
from pydantic import BaseModel
import matplotlib.pyplot as plt
import numpy as np
import cv2
import json


app = Flask(__name__,template_folder="")
def encode_image(image):
    _, encoded_image = cv2.imencode(".jpg", image)
    return "data:image/jpeg;base64," + base64.b64encode(encoded_image).decode()

# decode base64 string to image
def decode_image(image_string):
    encoded_data = image_string.split(',')[1]
    nparr = np.frombuffer(base64.b64decode(encoded_data), np.uint8)
    return cv2.imdecode(nparr, cv2.IMREAD_COLOR)

url = "http://localhost:8000"

@app.route('/')
def template():
    pdf_file_path = 'Coop101.pdf'

# Use the os.path.basename() function to get the filename
    pdf_file_name = os.path.basename(pdf_file_path)
    payload = {
        "image": pdf_file_name,
      
    }
    response = requests.post(f"{url}/process-file", json=payload)

    data = json.loads(response.content)

    processed_image_string = data["name"]

    return render_template("\Lab_jenkins_dockercompose\web\frontend\document.html", mycontent="processed_image_string")    

if __name__ == '__main__':
    app.run(host="0.0.0.0",port="8081")
```
